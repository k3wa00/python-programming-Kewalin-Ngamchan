Jag har valt att undersöka sambandet mellan personlig spendering, undervisningsutgifter och andelen elitstudenter. Jag har fokuserat på både andelen nya studenter från den översta 10 % och den översta 25 %.

För att göra detta har jag valt att använda korrelationsmatrisen. Tolkning av data kommer att vara: 

1: Positiv korrelation, dvs när en variabel ökar så ökar den andra också. 
-1: Negativ korrelation, dvs när en variabel ökar så minskar den andra. 
0: Ingen korrelation.

Expend och Top10perc visar en korrelation på 0.6609 vilket indikerar på en måttlig till stark korrelation mellan undervisningsutgifter och andelen studentender från översta 10%. Detta betyder att det finns en koppling mellan högre undervisningsutgifter till en höfre andel av elitstudenterna. 

När man kollar på korrelationen mellan Expend och Top25perc är det något mindre korrelation, 0.5274. Detta betyder ändå att det finns en måttlig positiv korrelation mellan undervisningsutgifter och andelen studenter från översta 25%. 

Sammanfattningsvis framgår det att de institutioner som tenderar att investera mer i undervisning också har en högre andel studenter som tillhör elitkategorin. Detta kan bero på flera faktorer, inklusive att dessa institutioner kan locka eller behålla mer kompetenta studenter. Det kan också indikera att resurser och stöd spelar en roll i studenternas framgång.

När man analyserar sambandet mellan personlig spendering och undervisningsutgifter framgår det en svag negativ korrelation, med ett värde på -0.0979. Dessutom är korrelationerna mellan personlig spendering och andelen elitstudenter svaga negativa, med värden på -0.0933 för den översta 10 % och -0.0808 för den översta 25 %. Trots att dessa korrelationer är svaga, indikerar de att högre personliga utgifter kan korrelera med lägre undervisningsutgifter.

Avslutningsvis är det viktigt att notera att dessa korrelationsvärden är svaga, vilket tyder på att det inte finns något starkt samband mellan personliga utgifter och undervisningsutgifter.
